# Problem 6: Text Classification

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

import numpy as np
```

```python
newsData = fetch_20newsgroups(subset='all')
len(newsData.data)
```

```python
XtrainData, XtestData, YtrainData, YtestData  = train_test_split(newsData.data, newsData.target, train_size=0.8, shuffle=True)
```

```python
vectorizer = CountVectorizer()

trainFeatureCount=vectorizer.fit_transform(XtrainData)
print(trainFeatureCount.shape)

testFeatureCount=vectorizer.transform(XtestData)
print(testFeatureCount.shape)
```

## Here by doing ‘vectorizer.fit_transform(newsData)’, we are learning the vocabulary dictionary and it returns a Document-Term matrix. [n_samples, n_features].

### It counts total number of samples and create a vocabulary list (I call them features) and returns #count of words: We can call this methodology as Bag of Words, layman: words count in a bag!

```python
vectorizer.get_feature_names();
len(vectorizer.get_feature_names())
```

```python
tfIdf = TfidfTransformer()

trainTfIdfData = tfIdf.fit_transform(trainFeatureCount)
print(trainTfIdfData.shape)

testTfIdfData = tfIdf.transform(testFeatureCount)
print(testTfIdfData.shape)
```

```python
classifier = MultinomialNB()
mlNetwork=classifier.fit(trainTfIdfData, YtrainData)
```

```python
result=mlNetwork.predict(testTfIdfData)
```

```python
accuracy = np.mean(result == YtestData)
print(accuracy)
```

```python
for doc, category in zip(YtestData, result):
...     print('%r => %s' % (doc, newsData.target_names[category]))
```

```python
accuracy_score(YtestData, result)
```

```python
f1_score(YtestData, result, average='macro')
```
